```python
mess_data = []
for messages in voices.messages:
  mess_data.append(messages)
```

```python
mess = pd.DataFrame(mess_data)
```

```python
mess['media_type'].value_counts()
```

```python
voice = mess[(mess['media_type'] == 'video_message') | (mess['media_type'] == 'voice_message')][['from','duration_seconds', 'media_type']]
```

```python
voice[voice['media_type'] == 'video_message']
```

```python
mess_from_vas = voice[voice['from'] == 'Прекрасная и Премудрая']['from'].count()
```

```python
mess_from_me = voice[voice['from'] == 'Andrew Nuzhdin']['from'].count()
```

```python
mess_total = len(voice)
```

```python
voices_from_vas = voice[(voice['from'] == 'Прекрасная и Премудрая') & (voice['media_type'] == 'voice_message')]['from'].count()
```

```python
voices_from_me = voice[(voice['from'] == 'Andrew Nuzhdin') & (voice['media_type'] == 'voice_message')]['from'].count()
```

```python
circles_from_vas = voice[(voice['from'] == 'Прекрасная и Премудрая') & (voice['media_type'] == 'video_message')]['from'].count()
```

```python
circles_from_me = voice[(voice['from'] == 'Andrew Nuzhdin') & (voice['media_type'] == 'video_message')]['from'].count()
```

```python
vas_ratio = mess_from_vas / mess_total * 100
```

```python
my_ratio = mess_from_me / mess_total * 100
```

```python
voices_duration = voice['duration_seconds'].sum()
```

```python
average_duration = voices_duration / mess_total
```

```python
vas_duration = voice[voice['from'] == 'Прекрасная и Премудрая']['duration_seconds'].sum()
```

```python
my_duration = voice[voice['from'] == 'Andrew Nuzhdin']['duration_seconds'].sum()
```

```python
vas_voices_duration = voice[(voice['from'] == 'Прекрасная и Премудрая') & (voice['media_type'] == 'voice_message')]['duration_seconds'].sum()
```

```python
my_voices_duration = voice[(voice['from'] == 'Andrew Nuzhdin') & (voice['media_type'] == 'voice_message')]['duration_seconds'].sum()
```

```python
vas_circles_duration = voice[(voice['from'] == 'Прекрасная и Премудрая') & (voice['media_type'] == 'video_message')]['duration_seconds'].sum()
```

```python
my_circles_duration = voice[(voice['from'] == 'Andrew Nuzhdin') & (voice['media_type'] == 'video_message')]['duration_seconds'].sum()
```

```python
vas_duration_ratio = vas_duration / voices_duration * 100
```

```python
my_duration_ratio = my_duration / voices_duration * 100
```

```python
vas_longest = voice[voice['from'] == 'Прекрасная и Премудрая']['duration_seconds'].max()
```

```python
my_longest = voice[voice['from'] == 'Andrew Nuzhdin']['duration_seconds'].max()
```

```python
vas_shortest = voice[voice['from'] == 'Прекрасная и Премудрая']['duration_seconds'].min()
```

```python
my_shortest = voice[voice['from'] == 'Andrew Nuzhdin']['duration_seconds'].min()
```

```python
import time
import datetime
voices_data = {'Vasilisa': [round(mess_total), round(mess_from_vas), round(abs(mess_from_vas - mess_from_me), 2), round(vas_ratio, 2), round(voices_from_vas), round(circles_from_vas), str(datetime.timedelta(seconds = voices_duration)), time.strftime("%M:%S", time.gmtime(average_duration)),
                          str(datetime.timedelta(seconds = vas_duration)), round(vas_duration_ratio, 2), str(datetime.timedelta(seconds = vas_voices_duration)), str(datetime.timedelta(seconds = vas_circles_duration)), time.strftime("%H:%M:%S", time.gmtime(vas_longest)), time.strftime("%M:%S", time.gmtime(vas_shortest))],
               'Andrey': [round(mess_total), round(mess_from_me), round(abs(mess_from_vas - mess_from_me), 2), round(my_ratio, 2), round(voices_from_me), round(circles_from_me), str(datetime.timedelta(seconds = voices_duration)), time.strftime("%M:%S", time.gmtime(average_duration)),
                          str(datetime.timedelta(seconds = my_duration)), round(my_duration_ratio, 2), str(datetime.timedelta(seconds = my_voices_duration)), str(datetime.timedelta(seconds = my_circles_duration)), time.strftime("%H:%M:%S", time.gmtime(my_longest)), time.strftime("%M:%S", time.gmtime(my_shortest))]}

stats = pd.DataFrame(voices_data, index = ['Total number of voices and circles', 'Number of voices and circles', 'Difference', "Number's ratio, %", 'Number of voices', 'Number of circles', "Total duration", "Average duration",
                                           'Duration', "Duration's ratio, %", "Voices' duration", "Circles' duration", 'Longest', 'Shortest'])
```
