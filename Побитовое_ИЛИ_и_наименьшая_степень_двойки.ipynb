## Определить, будет ли являться побитовое ИЛИ чисел в массиве чётным или нечётным числом.

Функция *parity_of_bitwise_or* принимает на вход массив целых чисел.

Для каждого числа из массива выполняется операция побитового ИЛИ с текущим значением, чтобы посчитать побитовое ИЛИ всех чисел в заданном массиве.

Затем определяется чётность полученного значения.

После определения функции вводятся длина массива чисел и сами числа.

В конце выдаётся результат: будет ли побитовое ИЛИ чисел в заданном массиве чётным или нечётным числом.

Внутри функции выполняется один проход по массиву длиной $n$.
Операции побитового ИЛИ и присваивания для каждого элемента выполняются за константное время $O(1)$.
Таким образом, общая временная сложность функции составляет $O(n)$, где $n$ — количество элементов в массиве.



````python
def parity_of_bitwise_or(arr):
  result = 0
  for num in arr:
    result |= num

  if result % 2 == 0:
    return('Побитовое ИЛИ чётное число')
  else:
    return('Побитовое ИЛИ нечётное число')

array = []
n = int(input('Введите длину массива: '))
for idx in range(n):
  array.append(int(input('Введите число в массиве: ')))
print(f'Массив чисел: {array}')
print(parity_of_bitwise_or(array))
````

## Определить наименьшую степень двойки, при возведении в которую получится значение большее или равное сумме чисел в массиве.

Для решения нам понадобятся функции *log2*, *ceil* из библиотеки *math*.

Функция *min_power_of_two* принимает на вход массив целых чисел.

Находится сумма чисел, входящих в массив.

Если сумма получается неположительной, то в этом случае будет выводится минус бесконечность, так как в пределе $2^n = 0$ при $n → -\infty$. В противном, для определения наименьшей степени двойки можно посчитать значение двоичного логарифма полученной суммы с помощью функции *log2* и округлить его в большую сторону с помощью функции *ceil*.

После определения функции вводятся длина массива чисел и сами числа.

В конце выдаётся результат: наименьшую степень двойки, при возведении в которую получится значение большее или равное сумме чисел в массиве.

Вычисление суммы требует одного прохода по всем элементам массива длиной $n$, то есть $O(n)$ операций сложения.

```ceil(log2(sum))``` обычно выполняется за константное время $O(1)$.
Таким образом, общая временная сложность составляет $O(n)$, где $n$ — количество элементов в массиве.


````python
from math import log2, ceil

def min_power_of_two(arr):
  sum = 0
  for num in arr:
    sum += num

  if sum <= 0:
    return 'минус бесконечность'
  else:
    return ceil(log2(sum))

array = []
n = int(input('Введите длину массива: '))
for idx in range(n):
  array.append(int(input('Введите число в массиве: ')))
print(f'Массив чисел: {array}')
print(f'Наименьшая степень двойки: {min_power_of_two(array)}')
````

Сумму чисел в массиве можно посчитать с использованием функции *sum* из библиотеки *numpy*.

Общая временная сложность также будет равна $O(n)$.

````python
from math import log2, ceil
import numpy as np

def min_power_of_two(arr):
  sum = np.sum(arr)

  if sum <= 0:
    return 'минус бесконечность'
  else:
    return ceil(log2(sum))

array = []
n = int(input('Введите длину массива: '))
for idx in range(n):
  array.append(int(input('Введите число в массиве: ')))
print(f'Массив чисел: {array}')
print(f'Наименьшая степень двойки: {min_power_of_two(array)}')
````
