## Задача

Дан многоугольник на плоскости, заданный  координатами своих вершин. Найти координаты вершин многоугольника, полученные поворотом каждой точки  на заданный угол вокруг начала координат. Вычислить средние значения координат по оси *ОХ* и *ОУ* повернутого многоугольника.

Снимок экрана 2026-02-18 в 21.00.12.png

**Входные данные**:

* количество вершин многоугольника (целое число);
* список вершин многоугольника, каждая вершина на отдельной строке, строка состоит из двух чисел через пробел, определяющие координаты вершины по оси *ОХ* и *OY* (целые числа);
*угол поворота в градусах (целое число).

**Выходные данные**:

среднее значение абсцисс (координат *x*) и ординат(координат *y*) точек повернутого многоугольника.

**Sample Input**:
```
3
-7 -8
-11 4
-9 5
45
```
**Sample Output**:
```
avg_x =  -6.60, avg_y =  -6.13
```

Создадим двухмерный массив из вершин многоугольника (количество строк - количество вершин, количество столбцов 2, для координат $x$ и $у$ точки).

Составим матрицу поворота вокруг центра координат ($\phi$ - значение угла поворота в радианах):

$$
rotate =
\begin{pmatrix}
cos(\phi) & sin(\phi) \\
- sin(\phi) & cos(\phi)
\end{pmatrix}
$$

Умножим массив вершин на матрицу поворота. Полученная матрица - координаты вершин многоугольника после поворота.

Найдём среднее значение первого столбца (**avg_x**) и среднее значение второго (**avg_y**).

````python
from math import sin, cos, radians
import numpy as np
````

````python
#n = int(input())
````

````python
n = 3
````

````python
#coords_array = []
#coords = input()
#coord_str = coords.split()
#coords_array.append(coord_str)
````

````python
coords_array = [['-7', '-8'], ['-11', '4'], ['-9', '5'], ['45']]
````

````python
phi = int(coords_array[n][-1])
phi_rad = radians(phi)
coords_array = coords_array[:n][:]
````

````python
coords_array = np.array(coords_array, dtype=int)
````

````python
rotate = np.array([[cos(phi_rad), sin(phi_rad)], [-sin(phi_rad), cos(phi_rad)]])
````

````python
coords_after_rotation = np.dot(coords_array, rotate)
````

````python
avg_x = np.mean(coords_after_rotation[:, 0])
avg_y = np.mean(coords_after_rotation[:, 1])
````

````python
print(f'среднее значение абсцисс = {avg_x:.2f}, среднее значение ординат = {avg_y:.2f}')
````
