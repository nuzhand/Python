```python
def multiply(a, b):

    if not b:
        return 0

    c = multiply(a, b >> 1)
    print('b =', b)
    print('b >> 1:', b >> 1)
    print('b & 1:', b & 1)
    print('c =', c)

    return a + (c << 1) if b & 1 else c << 1
```

```python
def length(x):
    n = 0
    if x <= 1: # Пришлось отдельно обрабатывать 0 и 1, т.к. битовый сдвиг дает 0
        return 1
    while x > 1:
        x >>= 1
        n += 1
    return n
```

```python
def two_halfs(a, n):
    x = a
    half_n = n >> 1
    x_l = a >> (half_n)
    x_r = a & ((1 << half_n) - 1)
    return x_l, x_r
```

```python
def karatsuba(x, y):
    n = max(length(x), length(y))
    if n == 1:
        return x * y
    half_n = n >> 1
    x_l, x_r = two_halfs(x, n)
    y_l, y_r = two_halfs(y, n)
    p1 = karatsuba(x_l, y_l)
    p2 = karatsuba(x_r, y_r)
    p3 = karatsuba(x_l + x_r, y_l + y_r)
    return (p1 << 2 * half_n) + ((p3 - p1 - p2) << half_n) + p2
```
